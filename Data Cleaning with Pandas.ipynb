## Question 2 - Clean up the data and make it more useful
    
</div>


```python
#Step 1 check for missing values:

df.isnull().sum()
```

```python
#Step 2 - fix the missing values: set the Genre NaN records = "NA"  
#tip - the Genre is datatype object (the example in the PowerPoint used float), so you will need to modify 
default_Genre = 'Others'
df['Genre'] = df['Genre'].fillna(default_Genre).astype(object)

df.isnull().sum()
```

```python
#Step 3 - Recheck for missing values, are they all Zero?

df.isnull().sum()
```

```python
#Step 4 - the State_Abbr field has Texas spelled out, replace the value with TX.

df['State_Abbr']= df['State_Abbr'].replace('Texas','TX')
df
```

```python
#Step 5 - Convert the Performance_Date from datatype object to datetime

df['Performance_Date'] = pd.to_datetime(df['Performance_Date'])


#verify that it has changed using  "dtypes" 

df.dtypes
```

<div style="background-color: #9DC9D5; padding: 10px;">
    
## Question 3 - Create new columns 
</div>


```python
#Step 1 - Create a new column for the 'Year_Qtr' column by formatting 'Performance_Date' with the to_period function

df['Performance_Month'] = df['Performance_Date'].dt.month
df['Year_Qtr'] = df['Performance_Date'].dt.to_period('Q')
df
```

```python
#Step 2 - Create a new column of your choice using either the bin_edges and bin lables approach

# Create a new column using bin_edges and bin lables
# create the bin for lables
bins = [0, 5000, 10000, 15000, 20000, 25000, float('inf')]

# defining the lable for each bin
labels = ['0 to 5,000', '5,000 to 10,000', '10,000 to 15,000', '15,000 to 20,000', '20,000 to 25,000', '25000 plus']

#create TotalRev_Buckets column
df['TotalRev_Buckets'] = pd.cut(df['TotalRev'], bins=bins, labels=labels, right=False)
df
```

```python
#Step 2 - Create a new column of your choice using the lambda function

df['Genre_Grouper'] = df['Genre'].apply(lambda x: 'Classical' if x == 'Classical' else 'Other' )
df
```
