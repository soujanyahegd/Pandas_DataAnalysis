```python
# Create a blank DataFrame named df_zip_info
df_zip_info = pd.DataFrame()

# Load data from the JSON Zip Code file into the DataFrame 
df_zip_info = pd.read_json("Zip_Code.json", lines=True)
```

```python
#inspect your dataframe

df_zip_info.info()
```

```python
df_zip_info.dtypes
```

```python
df_zip_info.describe(include='all')
```

```python
default_Household_Income = '0'
df_zip_info['Median_Household_Income'] = df_zip_info['Median_Household_Income'].fillna(default_Household_Income).astype(object)

df_zip_info.isnull().sum()
```

```python
default_Unemployed_Population = '0'
df_zip_info['Unemployed_Population'] = df_zip_info['Unemployed_Population'].fillna(default_Unemployed_Population).astype(object)

default_Below_Poverty_Level = '0'
df_zip_info['Below_Poverty_Level'] = df_zip_info['Below_Poverty_Level'].fillna(default_Below_Poverty_Level).astype(object)

default_Gini_Index = '0'
df_zip_info['Gini_Index'] = df_zip_info['Gini_Index'].fillna(default_Gini_Index).astype(object)


df_zip_info.isnull().sum()
```

```python
# Perform the join between df_performances and df_zip_info dataframes
df_performance_zip_stats = pd.merge(df_performances, df_zip_info,
                                    left_on='ZIP_CODE',
                                    right_on='zip_code_tabulation_area',
                                    how='inner')
df_performance_zip_stats
```

```python
#Create a new column of your choice using either the bin_edges and bin lables approach

# Create a new column using bin_edges and bin lables
# create the bin for lables
bins = [0, 3000000, 6000000, 9000000,float('inf')]

# defining the lable for each bin
labels = ['0 to 3,000,000', '3,000,000 to 6,000,000', '6,000,000 to 9,000,000', '9,000,000 plus']

#create TotalRev_Buckets column
df_performance_zip_stats['TotalRev_Buckets'] = pd.cut(df_performance_zip_stats['Total_Revenue_Amount'], bins=bins, labels=labels, right=False)
df_performance_zip_stats
```

```python
totalRev_Buckets_sum = df_performance_zip_stats.groupby('TotalRev_Buckets').size().reset_index(name = 'Record_Count')

totalRev_Buckets_sum
```

```python
bins = [0, 10000, 50000, 100000, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High']
df_performance_zip_stats['Average_Revenue_Bin'] = pd.cut(df_performance_zip_stats['Average_Revenue_Amount'], bins=bins, labels=labels)

df_performance_zip_stats
```

```python
bins = [0, 5, 15, 30, float('inf')]
labels = ['Rare', 'Occasional', 'Frequent', 'Very Frequent']
df_performance_zip_stats['Performance_Bin'] = pd.cut(df_performance_zip_stats['Unique_Performances'], bins=bins, labels=labels)

df_performance_zip_stats
```

```python
regions = {
    'NY': 'Northeast', 'CA': 'West', 'TX': 'South', 'FL': 'South', 'PA': 'Northeast', 'IL': 'Midwest', 
    'OH': 'Midwest', 'GA': 'South', 'NC': 'South', 'MI': 'Midwest', 'NJ': 'Northeast', 'VA': 'South', 
    'WA': 'West', 'AZ': 'West', 'MA': 'Northeast', 'TN': 'South', 'IN': 'Midwest', 'MO': 'Midwest', 
    'MD': 'Northeast', 'WI': 'Midwest', 'CO': 'West', 'MN': 'Midwest', 'AL': 'South', 'LA': 'South', 
    'KY': 'South', 'OR': 'West', 'OK': 'South', 'SC': 'South', 'KS': 'Midwest', 'NE': 'Midwest', 
    'UT': 'West', 'IA': 'Midwest', 'NV': 'West', 'CT': 'Northeast', 'MS': 'South', 'AR': 'South', 
    'NM': 'West', 'ID': 'West', 'HI': 'West', 'MT': 'West', 'WY': 'West', 'DE': 'Northeast', 'AK': 'West', 
    'SD': 'Midwest', 'ND': 'Midwest', 'VT': 'Northeast', 'RI': 'Northeast', 'ME': 'Northeast', 'DC': 'Northeast', 
    'PR': 'South'
}

# Assuming 'df' is your DataFrame and 'STATE_ABBR' is the column with state abbreviations
df_performance_zip_stats['Region'] = df_performance_zip_stats['STATE_ABBR'].map(regions)
df_performance_zip_stats
```

```python
# Create the pivot table
pivot_table = df_performance_zip_stats.pivot_table(index=['Region'],
                                    values=['Average_Revenue_Amount',
                                            'Unique_Performances',
                                            'Total_Population'],
                                    aggfunc={'Average_Revenue_Amount':'sum', 
                                            'Unique_Performances': 'sum',
                                             'Total_Population': 'sum'})

pivot_table['Average_Revenue_Per_Performance'] = pivot_table['Average_Revenue_Amount'] / pivot_table['Unique_Performances']
pivot_table


# Renaming the columns
pivot_table.columns = ['Total Average Revenue', 'Total Population', 'Unique Performance','Avg Revenue per Performance']
pivot_table

# Sort the pivot table
pivot_table_sorted = pivot_table.sort_values(by='Total Average Revenue', ascending=False)
pivot_table_sorted


# Format columns 
pivot_table['Total Average Revenue'] = pivot_table['Total Average Revenue'].apply('${:,.2f}'.format)
pivot_table['Avg Revenue per Performance'] = pivot_table['Avg Revenue per Performance'].apply('${:,.2f}'.format)
pivot_table['Total Population'] = pivot_table['Total Population'].astype(int).map('{:,}'.format)
pivot_table['Unique Performance'] = pivot_table['Unique Performance'].astype(int).map('{:,}'.format)

pivot_table
```

```python
styled_pivot = pivot_table.style.set_table_styles(
    {
        'Total Average Revenue': [{'selector': 'th', 'props': 'background-color: lightblue; color: black;'}],
        'Total Population': [{'selector': 'th', 'props': 'background-color: lightblue; color: black;'}],
        'Unique Performance': [{'selector': 'th', 'props': 'background-color: lightblue; color: black;'}],
        'Avg Revenue per Performance': [{'selector': 'th', 'props': 'background-color: lightblue; color: black;'}]    
    }
)

styled_pivot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the DataFrame with your data
data = {
    'Region': ['Midwest', 'Northeast', 'South', 'West'],
    'Total Avg Revenue ($)': [236574.96, 270109.53, 293921.72, 178007.28],
    'Total Population': [625277, 941205, 608772, 611072],
    'Unique Performance': [15756, 16833, 19145, 11747],
    'Avg Revenue/Performance ($)': [15.01, 16.05, 15.35, 15.15]
}
df = pd.DataFrame(data)

# Set the 'Region' as the index to display regions as rows
df.set_index('Region', inplace=True)

#Plot the heatmap
plt.figure(figsize=(18, 10))  # Adjust the size as needed
sns.heatmap(df, annot=True, cmap='YlOrBr', fmt=".2f", linewidths=0.5, cbar_kws={'label': 'Color Code with Numbers'})

# Adding title and labels
plt.title('Heatmap of Metrics by Region', fontsize=25, pad=25)
plt.xlabel('Metrics', fontsize=20)
plt.ylabel('Region', fontsize=20)

# Change the size of tick labels
plt.xticks(fontsize=12)  # Change x-axis tick label size
plt.yticks(fontsize=12)  # Change y-axis tick label size

# Show the plot
plt.show()
```
